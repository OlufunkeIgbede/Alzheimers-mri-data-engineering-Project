# Bronze Layer


Define Schema as STRING

```python
from pyspark.sql.types import StructType,StructField, StringType

crosssectional_schema = StructType([
  StructField("ID", StringType(), True),
  StructField("M/F", StringType(), True),
  StructField("Hand", StringType(), True),
  StructField("Age", StringType(), True),
  StructField("Educ", StringType(), True),
  StructField("SES", StringType(), True),
  StructField("MMSE", StringType(), True),
  StructField("CDR", StringType(), True),
  StructField("eTIV", StringType(), True),
  StructField("nWBV", StringType(), True),
  StructField("ASF", StringType(), True),
  StructField("Delay", StringType(), True),
])

Longitudinal_schema = StructType([
  StructField("SubjectID", StringType(), True),
  StructField("MRIID", StringType(), True),
  StructField("Group", StringType(), True),
  StructField("Visit", StringType(), True),  
  StructField(" MRDelay", StringType(), True),
  StructField("M/F", StringType(), True),
  StructField("Hand", StringType(), True),
  StructField("Age", StringType(), True),
  StructField("Educ", StringType(), True),
  StructField("SES", StringType(), True),
  StructField("MMSE", StringType(), True),
  StructField("CDR", StringType(), True),
  StructField("eTIV", StringType(), True),
  StructField("nWBV", StringType(), True),
  StructField("ASF", StringType(), True)
])
   
  
  
```

Read Raw file as csv

```python
crossscetionalschemadf = spark.read.format("csv").option("header", "true").schema(crosssectional_schema).load("/Volumes/healthdataproject/rawdata/raw_files/oasis_cross-sectional.csv")
Longitudinaldf = spark.read.format("csv").option("header", "true").schema(Longitudinal_schema).load("/Volumes/healthdataproject/rawdata/raw_files/oasis_longitudinal.csv")
```

Display Dataset

```python
display(crossscetionalschemadf)
display(Longitudinaldf)
```

Add Injestion Metadata

```python
from pyspark.sql.functions import current_timestamp

crossscetionalschemadf = crossscetionalschemadf.withColumn("ingesttimestamp", current_timestamp())
Longitudinaldf = Longitudinaldf.withColumn("ingesttimestamp", current_timestamp())
```

```python
display(crossscetionalschemadf)
display(Longitudinaldf)
```

```python
crossscetionalschemadf.write.format("delta").mode("overwrite").saveAsTable("healthdataproject.bronze.bronze_crosssectional")
Longitudinaldf.write.format("delta").mode("overwrite").saveAsTable("healthdataproject.bronze.bronze_longitudinal")
```

```python
Longitudinaldf = Longitudinaldf.toDF(*[c.strip().replace(" ", "_") for c in Longitudinaldf.columns])
```

```python
crossscetionalschemadf.write.format("delta").mode("overwrite").saveAsTable("healthdataproject.bronze.bronze_crosssectional")
Longitudinaldf.write.format("delta").mode("overwrite").saveAsTable("healthdataproject.bronze.bronze_longitudinal")
```

Load Bronze crosssection

```python
bronze_crossectional_df = spark.read.table("healthdataproject.bronze.bronze_crosssectional")
display(bronze_crossectional_df)
bronze_longitudinal_df = spark.read.table("healthdataproject.bronze.bronze_longitudinal")
display(bronze_longitudinal_df)
```

```python
bronze_longitudinal_df.select("M/F").distinct().display()
```

Remove Duplicate values

```python
silver_crossectional_df = bronze_crossectional_df.withColumnRenamed("M/F", "Sex")
silver_longitudinal_df = bronze_longitudinal_df.withColumnRenamed("M/F", "Sex")
```

Remove duplicate values

```python
from pyspark.sql.functions import count, countDistinct
silver_crossectional_df.select(count("ID"), countDistinct("ID")).display()
silver_longitudinal_df.select(count("SubjectID"), countDistinct("SubjectID")).display()

```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, desc

window_spec = Window.partitionBy("SubjectID").orderBy(desc("ingesttimestamp"))
silver_longitudinal_df = silver_longitudinal_df.withColumn("row_number", row_number().over(window_spec)).filter("row_number = 1").drop("row_number")

silver_longitudinal_df.display()

```

```python
silver_longitudinal_df.write.format("delta").mode("overwrite").saveAsTable("healthdataproject.silver.silver_longitudinal")
silver_crossectional_df.write.format("delta").mode("overwrite").saveAsTable("healthdataproject.silver.silver_crosssectional")
    
```
